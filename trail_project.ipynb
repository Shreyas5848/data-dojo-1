```python
from datadojo import create_dojo
dojo = create_dojo(educational_mode = True)
```

```python
from datadojo.dojo_api import Domain, DifficultyLevel
```

```python
projects = dojo.list_projects(
    domain = Domain.ECOMMERCE,
    difficulty = DifficultyLevel.BEGINNER
)

if not projects:
    print("No projects found")
    exit()

for p in projects:
    print(f"ID: {p.id}, Name: {p.name}")

selected_project_id = projects[0].id
```

```python
student_id = "Shreyas"
project = dojo.load_project(selected_project_id)
print(f"You have started : {project.info.name}")
```

```python
educational_interface = dojo.get_educational_interface()
explanation = educational_interface.get_concept_explanation("missing_values")
print(f"--{explanation['title']}--")
print(explanation['explanation'])
```

```python
data = project.dataset
print("\Original Data preview:")
print(data.head())
```

```python
data['age'] = data['age'].fillna(data['age'].mean())
```

```python
educational_interface.track_progress(
    student_id = student_id,
    project_id = selected_project_id,
    completed_step = "handled_missing_values",
    concepts_learned = ["missing_values", "imputation"]
)
```

```python
my_progress = project.get_progress(student_id)
print("\n--Your Progress--")
print(f"Completed Steps: {my_progress['completed_steps']}")
print(f"Learned Concepts: {my_progress['learned_concepts']}")
```
